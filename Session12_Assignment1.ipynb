```python
import numpy as np
import pandas as pd
from pandas import DataFrame, Series
import sqlite3 as db
from pandasql import sqldf
```

```python
pysqldf = lambda q: sqldf(q, globals())
```

```python
#loading data
sqladb=pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data")
```

```python
sqladb.head(5)
```

```python
#rename the columns
sqladb.columns =["age","workclass","fnlwgt","education","education_num","marital_status","occupation","relationship","race","sex","capital_gain","capital_loss","hours_per_week","native_country","income"]
sqladb.head(5)
```

```python
# Select 10 records from the adult sqladb
pysqldf("SELECT * FROM sqladb LIMIT 10")
```

```python
#Show the average hours per week of all men who are working in private sector
pysqldf('SELECT avg(hours_per_week) FROM sqladb WHERE workclass=" Private"')
```

```python
#Show the frequency table for education, occupation and relationship, separately

pysqldf('SELECT education, COUNT(*) FROM sqladb GROUP BY education')
```

```python
pysqldf('SELECT occupation, COUNT(*) FROM sqladb GROUP BY occupation')
```

```python
pysqldf('SELECT marital_status, COUNT(*) FROM sqladb GROUP BY marital_status')
```

```python
#Are there any people who are married, working in private sector and having a masters degree
pysqldf('SELECT COUNT(*) FROM sqladb WHERE marital_status=" Married-civ-spouse" AND workclass=" Private" AND education=" Masters"')
```

```python
#What is the average, minimum and maximum age group for people working in different sectors
pysqldf('SELECT occupation, min(age), max(age), avg(age) FROM sqladb GROUP BY occupation')
```

```python
#Calculate age distribution by country
pysqldf('SELECT native_country, age, COUNT(*) FROM sqladb GROUP BY age LIMIT 10')
```

```python
#Compute a new column as 'Net-Capital-Gain' from the two columns 'capital-gain' and 'capital-loss'
pysqldf('UPDATE sqladb SET capital_gain=capital_gain-capital_loss')
```
